```python
import json

with open('data/smarthome.json', 'r') as file:
    data = json.load(file)

len(data)
```

```python
with open('data/TSU/Annotation_MS-TEMBA_smarthome.json', 'r') as file:
    data2 = json.load(file)

len(data2)
```

```python
data2.keys() == data.keys()
```

```python
s = set()
for key in data.keys():
    for i in range(len(data[key]['actions'])):
        s.add(data[key]['actions'][i][0])
print(s)
print(len(s))
```

```python
c = 0
for key in data.keys():
    if len(data[key]['actions']) != len(data2[key]['actions']):
        print(key, len(data[key]['actions']), len(data2[key]['actions']))
        print(data[key]['actions'])
        print(data2[key]['actions'])
        c += 1
print(c, 'keys have different number of actions')
```

```python
from collections import defaultdict

d = defaultdict(int)
for key in data.keys():
    for i in range(len(data[key]['actions'])):
        d[data[key]['actions'][i][0]] += 1
print(d)
print(len(d))

# print sorted by value
print(sorted(d.items(), key=lambda x: x[1], reverse=True))
```

```python
d2 = defaultdict(int)

for key in data2.keys():
    for i in range(len(data2[key]['actions'])):
        d2[data2[key]['actions'][i][0]] += 1
print(d2)
print(len(d2))

# print sorted by value
print(sorted(d2.items(), key=lambda x: x[1], reverse=True))
```

```python
# unique subset values in data
subsets = set()
for key in data.keys():
    subsets.add(data[key]['subset'])

print(subsets)
```

```python
# match train and test splits from data into data2

for key in data.keys():
    if data[key]['subset'] == 'training':
        data2[key]['subset'] = 'training'
    elif data[key]['subset'] == 'testing':
        data2[key]['subset'] = 'testing'
```
