```python
import numpy as np
import os
import json

with open("data/TSU/Annotation_MS-TEMBA_smarthome.json", "r") as f:
    ann = json.load(f)

files = [
    "data/TSU/i3d_features/P02T01C06.npy",
    "data/TSU/i3d_features/P18T14C07.npy",
    "data/TSU/i3d_features/P25T17C02.npy",
]

for f in files:
    vid = os.path.splitext(os.path.basename(f))[0]
    x = np.load(f)

    non_zero = np.any(x != 0, axis=1)
    valid = np.where(non_zero)[0][-1] + 1 if np.any(non_zero) else 0
    trailing_zero_ok = np.all(x[valid:] == 0)

    duration = ann[vid]["duration"]
    ratio = valid / duration if duration > 0 else None
    max_end = max(a[2] for a in ann[vid]["actions"]) if ann[vid]["actions"] else None

    print(
        vid,
        "shape=", x.shape,
        "valid=", valid,
        "duration=", duration,
        "ratio=", round(ratio, 4) if ratio is not None else None,
        "max_end=", max_end,
        "tail_zero=", trailing_zero_ok
    )
```
